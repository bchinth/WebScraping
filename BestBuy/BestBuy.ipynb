# Part 1: Google Searches

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time

driver = webdriver.Chrome()

driver.implicitly_wait(10)
driver.set_script_timeout(120)
driver.set_page_load_timeout(10)

driver.get('https://google.com')

time.sleep(3)


input = driver.find_element(By.CSS_SELECTOR, 'textarea[title="Search"]')
input.send_keys('askew\n')

time.sleep(10)

driver.quit()
```

# Part 2: Best Buy

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service 
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time 

driver = webdriver.Chrome()

time.sleep(4)

driver.get('https://www.bestbuy.com/')

Dod = driver.find_element(By.XPATH, '//a[text()="Deal of the Day"]')
Dod.click()

time.sleep(3)

print(Dod)

# Method 1
countdown = driver.find_element(By.XPATH, '//div[@class="countdown-clock"]') # using div 
print(countdown.text)

print()

# Method 2
countdown = driver.find_element(By.XPATH, '//ul[@id="countdown"]') # using ul
print(countdown.text.split('\n'))

# Format the countdown timer
countdown_list = countdown.text.split('\n')  
countdown_string = ''.join(countdown_list)

print(f"\n Remaining Time: {countdown_string} \n")

time.sleep(3)

# Click on the actual product
Dod_ap = driver.find_element(By.XPATH, '//h1[@tabindex="-1"]')
Dod_ap.click()

time.sleep(3)

reviews = driver.find_element(By.XPATH, '//a[@data-track="stats: customerReviews"]') 
reviews.click()


# Print Reviews
print(reviews.text)

print(reviews)

# Save reviews html 
with open("bestbuy_deal_of_the_day.htm", 'w', encoding='utf-8') as file:
            file.write(driver.page_source)
            print(f"\n Saved bestbuy_deal_of_the_day.htm")

time.sleep(4)

driver.quit()
```

# Part 3: Lavazza Coffee

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service 
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time 

driver = webdriver.Chrome()

time.sleep(3)

driver.get('https://www.lavazzausa.com/')

time.sleep(4)

# Reject Cookie
cookie = driver.find_element(By.XPATH, '//div[text()="        REJECT ALL     "]')
cookie.click()

time.sleep(3)

# Click on Products

#products = driver.find_element(By.XPATH, '//a[text()="PRODUCTS"]')
#products.click()
driver.find_element(By.XPATH, '//a[text()="PRODUCTS"]').click() 
time.sleep(3)

# Click on Capsule Coffee Machines
driver.find_element(By.XPATH, '//a[text()="Capsule coffee machines"]').click()
time.sleep(3)


# Save Capsule coffee machines html 
with open("lavazza.htm", 'w', encoding='utf-8') as file:
            file.write(driver.page_source)
            print(f"\n Saved lavazza.htm")

time.sleep(4)


driver.quit()
```

```python
def readhtml(f):
	try:
		html = open(f, "r", encoding='utf-8').read()
		return(html)
	except Exception as ex:
		print('Error: ' + str(ex))

html = readhtml("lavazza.htm")
print(html)
```

```python
import os
from bs4 import BeautifulSoup

with open("lavazza.htm", 'r', encoding='utf-8') as file:
        
        # Parse HTML content 
        soup = BeautifulSoup(file, 'html.parser')

        # Extract the information
        product = soup.find('h2', class_='lvz-product-card__name').get_text()
        prices = soup.find('p', class_='lvz-product-card__pricing').get_text().split('\n')

        # Print the values
        print(f"Product: {product}")
        print(f"Original Price: {prices[1]}")
```
