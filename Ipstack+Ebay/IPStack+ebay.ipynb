## Part 2: eBay + Selenium

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains
import time

driver = webdriver.Chrome()

driver.implicitly_wait(10)
driver.set_script_timeout(120)
driver.set_page_load_timeout(10)

#driver.get('https://google.com')
driver.get('https://www.ebay.com/')


time.sleep(3)


input = driver.find_element(By.XPATH, "//input[@placeholder='Search for anything']")
input.send_keys('Cell Phones\n')
#time.sleep(10)
time.sleep(5)


time.sleep(5)

# filter network
filters = driver.find_element(By.XPATH, f"//span[contains(text(), 'Unlocked')]")
filters.click()
time.sleep(3)

# scroll after first filter
filters = driver.find_elements(By.XPATH, f"//span[contains(text(), 'Network')]/following::span[contains(text(), 'Unlocked')]")
if filters:
        # Scroll to the filter checkbox
        driver.execute_script("arguments[0].scrollIntoView(true);", filters[0])
        ActionChains(driver).move_to_element(filters[0]).perform()
time.sleep(3)


# Filter Brand
filters = driver.find_element(By.XPATH, f"//span[contains(text(), 'LG')]")
filters.click()
time.sleep(3)

# scroll after LG filter
filters = driver.find_elements(By.XPATH, f"//span[contains(text(), 'Brand')]/following::span[contains(text(), 'LG')]")
if filters:
        # Scroll to the filter checkbox
        driver.execute_script("arguments[0].scrollIntoView(true);", filters[0])
        ActionChains(driver).move_to_element(filters[0]).perform()
time.sleep(3)



# Filter Screen Size
filters = driver.find_element(By.XPATH, f"//span[contains(text(), '6 in or More')]")
filters.click()
time.sleep(3)


# scroll after scrren size filter
filters = driver.find_elements(By.XPATH, f"//span[contains(text(), 'Screen Size')]/following::span[contains(text(), '6 in or More')]")
if filters:
        # Scroll to the filter checkbox
        driver.execute_script("arguments[0].scrollIntoView(true);", filters[0])
        ActionChains(driver).move_to_element(filters[0]).perform()
time.sleep(3)

# Filter Storage Capacity
filters = driver.find_element(By.XPATH, f"//span[contains(text(), '128 GB')]")
filters.click()
time.sleep(3)

# Filter nStorage Capacity
filters = driver.find_element(By.XPATH, f"//span[contains(text(), 'Factory Unlocked')]")
filters.click()
time.sleep(3)

# filter Condition
filters = driver.find_element(By.XPATH, f"//span[contains(text(), 'Used')]")
filters.click()
time.sleep(3)

# save to html file
with open("unlocked-lg-128gb-used.html", "w", encoding="utf-8") as file:
    file.write(driver.page_source)


driver.quit()
```

```python
from bs4 import BeautifulSoup
import re

# open html file
with open("unlocked-lg-128gb-used.html", "r", encoding="utf-8") as file:
    soup = BeautifulSoup(file, "html.parser")
    #soup.prettify



# iterate through each item listing
for i in soup.select('.s-item'):
    # To extract title
    title = i.select_one('.s-item__title').text 
    
    # To extract seller info and percentage (rating)
    seller = i.select_one('.s-item__seller-info-text').text if i.select_one('.s-item__seller-info-text') else "N/A"
    
    # Extract seller rating - using regex
    rating = re.search(r'(\d+(\.\d+)?)%', seller)

    # If rating is there, extract the percentage
    if rating: seller_rating = rating.group(1)
    else: seller_rating = "N/A"

    # To extract the price
    price = i.select_one('.s-item__price').text 
    
    # Shipping information  
    shipping = i.find("span", class_="s-item__shipping s-item__logisticsCost").text if i.find("span", class_="s-item__shipping s-item__logisticsCost") else "Shipping information - Not available"
   
    # Returns information
    returns = i.find("span", class_="s-item__shipping s-item__logisticsCost").text if i.find("span", class_="s-item__shipping s-item__logisticsCost") else "Shipping information - Not available "
   

    # Number of bids 
    bids = i.find("span", class_="s-item__dynamic s-item__quantitySold").text if i.find("span", class_="s-item__dynamic s-item__quantitySold") else "Bidding Info - Not Available"
   
    

    print(f"Title: {title}")
    print(f"Seller: {seller}")
    print(f"Seller Rating (in %): {seller_rating}")
    print(f"Price: {price}")

    print(f"Shipping Information: {shipping}")
    print(f"Return Information: {returns}")

    print(f"Number of bids: {bids}")
    print("-------\n")  
```
