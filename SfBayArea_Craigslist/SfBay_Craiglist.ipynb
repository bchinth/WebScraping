 **Interact with the Page-Sorting:**

1. Sorting URL's


*   Sort: New to Old -

Original URL: https://sfbay.craigslist.org/search/zip#search=1~gallery~0~0

*   Sort: Old to New -

Sorting URL: https://sfbay.craigslist.org/search/zip?sort=dateoldest#search=1~gallery~0~0



2. Toggled the URL parameters from oldest to newest.
https://sfbay.craigslist.org/search/zip?sort=datenewest#search=1~gallery~0~0

3. Here we changed the parameters in the URL by appending the 'sort = ' tot he original URL. The browser sends a 'GET' request to the server with this new updated URL. Then server processes the request and responds with the listings sorted in the given order.

4. 'sort' is the variable in URL. 'dateoldest' and 'datenewest' are the values indicating the sorting order.

 **Interact with the Page-Pagination:**

1. Changes in URL - Pages


*  Page 1(0-120):  https://sfbay.craigslist.org/search/zip#search=1~gallery~0~0
*  Page 2(121 - 240):  https://sfbay.craigslist.org/search/zip#search=1~gallery~1~0
*  Page 3(241 - 360):  https://sfbay.craigslist.org/search/zip#search=1~gallery~3~0

2. The number after 'gallery~' represent the page number. 1 corresponds to the second page, and 3 corresponds to the thrird page.

3. The variable is the page number after 'gallery~'. The part 'gallery~1' and 'gallery~2' instructs the website to display next 120 set of listings. Each page is associated with a portion of overall result to display and by incrementing this number we move to the enxt page.

```python
import requests
from bs4 import BeautifulSoup

time.sleep(5)

def fetch_listings_urls(num_listings):
    base_url = 'https://sfbay.craigslist.org/search/zip'
    listings_urls = []

    num_pages = (num_listings - 1) // 120 + 1 #3 pages

    for page_number in range(num_pages):
        offset = (page_number * 120) + 1 #listings range in each page 1-120, 121-240,
        current_page_url = f'{base_url}#search=1~gallery~{page_number}~0' #current page URL

        response = requests.get(current_page_url)
        soup = BeautifulSoup(response.text, 'html.parser')
        current_page_listing_links = soup.select('li.cl-static-search-result a') #selecting class

        # append the URLs to the list
        for link in current_page_listing_links:
                listings_urls.append(link['href']) #collecting each listings link into a list

                # max listins < 250
                if len(listings_urls) >= num_listings:
                    return listings_urls

    return listings_urls

# first 250 listings URLs
num_listings = 250
listings_urls_set = fetch_listings_urls(num_listings)

# Print the result
for index, url in enumerate(listings_urls_set):
    print(f"{index + 1}. {url}")
```

```python
# To save links to html files
def save_html(listing_url, listing_id):
    response = requests.get(listing_url)

    try:
        # Save the HTML content to a file
        with open(f"{listing_id}.html", 'w', encoding='utf-8') as file:
            file.write(response.text)
        print(f"Saved {listing_id}.html")
    except Exception as ex:
        print('Error: ' + str(ex))
```

```python

# Fetch and save HTML content for each listing
for url in listings_urls_set:
    # Extract the listing ID from the URL
    listing_id = url.split('/')[-1].split('.')[0]
    save_html(url, listing_id)
```

```python
import os
from bs4 import BeautifulSoup

def extract_information_from_html(html_file_path):
    with open(html_file_path, 'r', encoding='utf-8') as file:
        # Parse the HTML content of the file
        soup = BeautifulSoup(file, 'html.parser')

        # Extract information
        title = soup.select_one('span.postingtitletext').get_text(strip=True)
        first_image_url = soup.select_one('img.swipe img')['src'] if soup.select_one('img.swipe img') else None
        description = soup.select_one('section#postingbody').get_text(strip=True)
        post_id = soup.find('p', class_='postinginfo').get_text(strip=True).split(":")[-1].strip()
        posted_date = soup.select_one('time')['datetime']
        last_updated_date = soup.select_one('p.postinginfo time')['datetime']

        # Print the extracted information
        print(f"Title: {title}")
        print(f"URL of first image: {first_image_url}")
        print(f"Description: {description}")
        print(f"Post ID: {post_id}")
        print(f"Posted Date: {posted_date}")
        print(f"Last Updated Date: {last_updated_date}")
        print("\n")

# Directory where HTML files are located
html_files_directory = "/content/"  # Change this to the directory where your HTML files are located

# Loop through HTML files in the specified directory
for filename in os.listdir(html_files_directory):
    # Check if the file ends with .html
    if filename.endswith(".html"):
        html_file_path = os.path.join(html_files_directory, filename)
        print(f"Extracting information from: {filename}")
        extract_information_from_html(html_file_path)
```

**4. Automating the Login Process**

```python
input = soup.select_one('#new_user input[name=authenticity_token]')
authenticity_token = input.get('value')
print(f"authenticity_token - {authenticity_token}")

# Always pause between two requests.
time.sleep(5) # 5s

# An open session carries the cookies and allows you to make post requests
session = requests.session()

res = session.post('https://theoldreader.com/users/sign_in',
                        data = {'utf8': '',
                                'authenticity_token': authenticity_token,
                                'user[login]': 'bchinth@ucdavis.edu',
                                'user[password]': 'Prime@999',
                                'commit': 'Sign In'
                               },
                        timeout = 20)


# This will get us the Cookies.
cookies = session.cookies.get_dict()
print(cookies)
```
